## 5. Test G-Code Visualization

```python
# Test creating and visualizing G-code
try:
    from promptplot.core import GCodeCommand, GCodeProgram
    
    # Create individual G-code commands to draw a square
    commands = [
        GCodeCommand(command="G28"),  # Home
        GCodeCommand(command="M5"),   # Pen up
        GCodeCommand(command="G1", x=0, y=0),    # Move to origin
        GCodeCommand(command="M3"),   # Pen down
        GCodeCommand(command="G1", x=50, y=0),   # Draw line to (50,0)
        GCodeCommand(command="G1", x=50, y=50),  # Draw line to (50,50)
        GCodeCommand(command="G1", x=0, y=50),   # Draw line to (0,50)
        GCodeCommand(command="G1", x=0, y=0),    # Draw line back to origin
        GCodeCommand(command="M5"),   # Pen up
    ]
    
    # Create a G-code program with the commands
    program = GCodeProgram(commands=commands)
    
    print(f"✓ Created G-code program with {len(program.commands)} commands")
    print("\nG-code preview:")
    for i, cmd in enumerate(program.commands[:4]):
        print(f"  {i+1}: {cmd.to_gcode()}")
    if len(program.commands) > 4:
        print(f"  ... and {len(program.commands) - 4} more commands")
    
    # Test program analysis methods
    bounds = program.get_bounds()
    if bounds:
        print(f"\nDrawing bounds: {bounds}")
    
    command_counts = program.count_by_command_type()
    print(f"Command counts: {command_counts}")
    
    # Test converting back to G-code string
    gcode_string = program.to_gcode()
    print(f"\nFull G-code output ({len(gcode_string)} characters):")
    print(gcode_string[:200] + "..." if len(gcode_string) > 200 else gcode_string)
    
    # Visualize the drawing path
    movement_commands = program.get_movement_commands()
    if movement_commands:
        x_coords = [cmd.x for cmd in movement_commands if cmd.x is not None]
        y_coords = [cmd.y for cmd in movement_commands if cmd.y is not None]
        
        if x_coords and y_coords:
            plt.figure(figsize=(8, 8))
            plt.plot(x_coords, y_coords, 'b-', linewidth=2, marker='o', markersize=4)
            plt.xlabel('X')
            plt.ylabel('Y')
            plt.title('G-code Path Visualization')
            plt.grid(True, alpha=0.3)
            plt.axis('equal')
            plt.show()
            print("✓ G-code path visualization created!")
        
except Exception as e:
    print(f"❌ G-code creation error: {e}")
    import traceback
    traceback.print_exc()
```
